```python
import pandas as pd
import numpy as np
import sklearn

#Загрузка и конкатенация данных
data11 = pd.read_csv('opel_corsa_01.cvs', delimiter=';')
data12 = pd.read_csv('opel_corsa_02.cvs', delimiter=';')
data21 = pd.read_csv('peugeot_207_01.cvs', delimiter=';')
data22 = pd.read_csv('peugeot_207_02.cvs', delimiter=';')
data = pd.concat([data11, data12, data21, data22], ignore_index=True)

#Присвоение признаков
x = data.values[::, 0:14]
y = data.values[::, 14:15]

#NaN, стандартизация, кодирование srt признаков (НЕ РАБОТАЕТ)
from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy="most_frequent")
x = imp.fit_transform(x)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(x)
x = scaler.fit_transform(x)

from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
y = enc.fit_transform(y)

#Разделение на поезда
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

print(x_train)
print(y_train)
```
